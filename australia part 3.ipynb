### Extracting features

```python
# Extracting features from the Engine column
df['Engine'] = df['Engine'].str.split(',').str[1].str.strip('L')

# Removin 'cyl' in cylinder col
df['CylindersinEngine'] =df['CylindersinEngine'].str.strip('cyl')

# Removing "/100 km"
df['FuelConsumption'] = df['FuelConsumption'].str.split('/').str[0].str.strip('/100 km')

# Removing "L" from fuel consumption column
df['FuelConsumption'] =df['FuelConsumption'].str.strip('L')

# Removing "Seats"from the seats column
df['Doors']=df['Doors'].str.strip('Doors')

# Removing "Doors" from the doors column
df['Seats']=df['Seats'].str.strip('Seats')

# Extracting the exterior and interior colors from ColourExtint
df['Extcolour'] = df['ColourExtInt'].str.split('/').str[0]
df['Intcolour'] = df['ColourExtInt'].str.split('/').str[1]

# Extracting state and city from Location
df['City'] = df['Location'].str.split(',').str[0] 
df['State'] = df['Location'].str.split(',').str[1]
```

# Handling misappropriate column values


```python
# Converting values to nan values
df['FuelType'] = df['FuelType'].replace('-', 'Unknown')
df['Price'] = df['Price'].replace('POA', np.nan)
df['Kilometres'] = df['Kilometres'].replace('- / -', np.nan)
df['Kilometres'] = df['Kilometres'].replace('-', np.nan)
df['Transmission'] = df['Transmission'].replace('-', np.nan)
df['FuelConsumption/100km'] = df['FuelConsumption/100km'].replace('-', np.nan)
df['FuelConsumption/100km'] = df['FuelConsumption/100km'].replace(r'^\s*$', np.nan, regex=True)
df['CylindersinEngine'] = df['CylindersinEngine'].replace('-', 'Unknown')
df['Extcolour'] = df['Extcolour'].replace(r'-', np.nan, regex=True)
df = df.drop(columns=['Intcolour'])
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Engine','Kilometres', 'Doors', 'Seats', 'Price', 'City', 'State', 'Extcolour','Transmission'], inplace=True)
```

# Assigning the right dtype for each column

```python
df['Year'] = pd.to_datetime(df['Year'].astype('Int64'), format='%Y')
df['Engine'] = pd.to_numeric(df['Engine'])
df['FuelConsumption/100km'] = pd.to_numeric(df['FuelConsumption/100km'])
df['CylindersinEngine'] = pd.to_numeric(df['CylindersinEngine'])
df['Seats'] = pd.to_numeric(df['Seats'])
df['Doors'] = pd.to_numeric(df['Doors'])
```

```python
median_fuel = df['FuelConsumption/100km'].median()
df['FuelConsumption/100km'] = df['FuelConsumption/100km'].fillna(median_fuel)
```

# Line plot
A line plot is best used to show trends, patterns, or changes in data over a continuous and ordered variable, most commonly time. It helps visualize how something changes over intervalsâ€”like years, months, or measurements.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Just use fixed, safe bin edges (make last one very large)
bins = [0, 200000, 400000, 600000, 800000, 1000000, 10000000]
labels = ['0-200k', '200k-400k', '400k-600k', '600k-800k', '800k-1M', '1M+']

# Bin the prices
df['Price_Bin'] = pd.cut(df['Price'], bins=bins, labels=labels, include_lowest= True)

# Group and get average fuel consumption
binned = df.groupby('Price_Bin')['Engine'].mean().reset_index()

# Plot the result
plt.figure(figsize=(10, 6))
plt.plot(binned['Price_Bin'], binned['Engine'], marker='o', color='green')
plt.xlabel('Price Range')
plt.ylabel('Average Engine capacity')
plt.title('Engine capacity by Price Range')
plt.grid(True)

plt.show()
```

```python
import plotly.express as px

bins = [0, 200000, 400000, 600000, 800000, 1000000, 10000000]
labels = ['0-200k', '200k-400k', '400k-600k', '600k-800k', '800k-1M', '1M+']

# Bin the prices
df['Price_Bin'] = pd.cut(df['Price'], bins=bins, labels=labels, include_lowest= True)

# Group and get average fuel consumption
binned = df.groupby('Price_Bin')['Engine'].mean().reset_index()
fig = px.line(binned, x='Price_Bin', y='Engine', markers=True, title='Average Engine Size by Price Range')

fig.update_traces(line_color='red')
fig.update_layout(
    xaxis_title='Price Range',
    yaxis_title='Average Engine Size',
    title_x=0.5
)
fig.show()
```

### Box Plots

A box plot is a statistical visualization that displays the five-number summary (minimum, Q1, median, Q3, maximum) with the box representing the middle 50% of data and whiskers extending to non-outlier extremes. Outliers appear as individual points beyond 1.5Ã—IQR from the box edges. Box plots enable rapid assessment of data distribution, quality issues, and cross-group comparisons in exploratory analysis. They effectively reveal skewness, variance differences, and potential transformation needs before statistical modeling.
```python
# Lets visualize the distribution of usedornew

fig = px.box(df, y='UsedOrNew', x='Price',color='UsedOrNew', title='Price Distribution: Demo, Used & New Cars')
fig.update_layout(
    xaxis_title='Price',
    yaxis_title='Car condition',
)
fig.update_traces(orientation='h')
fig.show()
```

* USED cars: Have the widest price range, many outliers, and a high maximum value (reaching over 600k).This suggests significant variability, some used cars are extremely expensive.

* DEMO cars: Prices are generally clustered around 100k, but a few high-end demo cars exist.Fewer outliers compared to used cars.

* NEW cars: Show a tighter, more compact distribution, implying more consistent pricing. Typically range from under 100k to about 150k.Also have a few outliers, but less spread than USED.

```python
# Visualizing the distribution of Transmission

#Count the occurrences of each transmission type
transmission_counts = df['Transmission'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 8))
colors = ['skyblue', 'lightgreen']
plt.bar(transmission_counts.index, transmission_counts.values, color= colors)

# Set custom labels
plt.xticks(ticks=range(len(transmission_counts)), labels=['Automatic', 'Manual'])

# Add titles and labels
plt.title('Distribution of Transmission Types')
plt.xlabel('Transmission Types')
plt.ylabel('Count')
plt.grid(True)

# Show the plot
plt.show()
```

* Automatic cars dominate the market, making up over 85% of listings (~12,000), indicating strong availability and preference. Manual cars are few (~1,600), comprising less than 15% of listings â€” possibly due to changing consumer trends or newer imports being mostly automatic.
* For buyers: More options exist for automatics.
* For sellers: Automatics face higher competition but likely meet higher demand. Manuals may have niche appeal but limited demand.

```python
fig = px.box(df, x='Transmission', y='Price', title='Transmission vs Price Distribution', color='Transmission')

fig.update_layout(
    xaxis_title='Transmission Type',
    yaxis_title='Price (Log Scale)',
    yaxis_type='log'  # log scale for better visibility
)
fig.show()
```

* Most cars have an Automatic transmission and also the median price of automatic cars (29.8K ) is higher that manual cars (18.98k).
* Both transmission types have outliers (dots above the whiskers), indicating some very high-priced listings in each group.
* The use of a logarithmic scale makes it clear that there are extremely low-priced manual cars (possibly very old or faulty models), which would be hidden on a normal scale. Automatic cars donâ€™t dip as low in price â€” suggesting a higher minimum market value.
